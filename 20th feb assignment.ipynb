```python
#Question 1:

GET: The Get method is mainly used at the client (Browser) side to send a request to a specified server to get certain data or resources. Using this method the server should only let us receive the data and not change its state. Hence it is only used to view something and not to change it. Get method is one of the most used HTTP method.

POST: The Post method is mainly used at the client (Browser) side to send data to a Specified server in order to create or rewrite a particular resource/data. This data sent to the server is stored in the request body of the HTTP request. Post method eventually leads to the creation of a new resource or updating an existing one. Due to this dynamic use, it is one of the most used HTTP methods. It is not one of the most secure methods because the data that is been sent is included in the body of the request and not in the URL.
```

```python
#Question 5:

from flask import*

app = Flask(__name__)

@app.route('/postman_data', methods = ['POST'])
def operation():
    if(request.method =='POST'):
        op = request.json['operator']
        num1 = int(request.json['num1'])
        num2 = int(request.json['num2'])
        if(op == 'add'):
            res = num1 + num2
            result = 'the sum of '+ str(num1) + 'and' + str(num2) + 'is' + str(res)
       
        if(op == 'subtract'):
            res = num1 - num2
            result = 'the Difference of '+ str(num1) + 'and' + str(num2) + 'is' + str(res)

        if(op == 'multiply'):
            res = num1 * num2
            result = 'the multiplication of '+ str(num1) + 'and' + str(num2) + 'is' + str(res)

            
        if(op=='divide'):
            res = num1/num2
            result = 'the divide of '+ str(num1) + 'and' + str(num2) + 'is' + str(res)

            
        return jsonify(result)
    
if __name__=="__main__":
    app.run(host="0.0.0.0")
            
```
